```python
import json
import numpy as np
import matplotlib.pyplot as plt
import torch
```

```python
file_path = "PACLITAXEL_g2s.json"

records = []
with open(file_path, "r") as file:
    data = json.load(file)  # 배열 형태로 로드

record_0 = data.get("0") 
record_1 = data.get("1") 
record_2 = data.get("2") 
record_3 = data.get("3") 
```

```python
drug_graph = torch.load("./model/input_full/0_drug_graph_dict.pt")
drug_subs = sorted(drug_graph['PACLITAXEL']['global_ids'])
drug_subs
```

```python
# Sensitive
# Cell line ACH-000648, Drug PACLITAXEL, Pathway 79
record = record_0

# 데이터 추출
cell_line = record['cell_line']
drug = record['drug']
pathway = record['pathway']

valid_gene_indices = record["valid_gene_indices"]
set_global_ids = record["set_global_ids"]
weights = np.array(record["weight"])

min_weight = weights.min()
max_weight = weights.max()

print("min_weight : ", min_weight)
print("max_weight : ", max_weight)

global_min_weight = -0.0158469807356596
global_max_weight = 0.13336576521396637

# 선택할 set_global_ids
selected_global_ids = drug_subs  # 선택된 drug_subs
selected_indices = [set_global_ids.index(gid) for gid in selected_global_ids]

# weights 배열에서 선택된 열만 필터링
filtered_weights = weights[:, selected_indices]

# 공통 스케일 설정
global_min_weight = -0.0158469807356596
global_max_weight = 0.13336576521396637

# 히트맵 생성
plt.figure(figsize=(10, 8))
plt.imshow(filtered_weights, cmap="coolwarm", aspect="auto", vmin=global_min_weight, vmax=global_max_weight)
plt.colorbar(label="Weight")  # 컬러바 추가

# 축 레이블 설정
plt.xticks(ticks=np.arange(len(selected_global_ids)), labels=selected_global_ids, rotation=90)
plt.yticks(ticks=np.arange(len(valid_gene_indices)), labels=valid_gene_indices)

# 제목 및 축 레이블 추가
plt.title(f"{cell_line}-{drug}-{pathway} Weight Heatmap (Filtered)")
plt.xlabel("Drug Substructures")
plt.ylabel("Gene Indices")

# 히트맵 표시
plt.tight_layout()
```
