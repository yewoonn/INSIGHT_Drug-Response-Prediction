```python
import json
import numpy as np
import matplotlib.pyplot as plt
import torch
```

```python
file_path = "PACLITAXEL_g2s.json"

records = []
with open(file_path, "r") as file:
    data = json.load(file)  # 배열 형태로 로드

record_0 = data.get("0") 
record_1 = data.get("1") 
record_2 = data.get("2") 
record_3 = data.get("3") 
```

```python
# Sensitive
# Cell line ACH-000648, Drug PACLITAXEL, Pathway 0
record = record_0

# 데이터 추출
cell_line = record['cell_line']
drug = record['drug']
pathway = record['pathway']

valid_gene_indices = record["valid_gene_indices"]
set_global_ids = record["set_global_ids"]
weights = np.array(record["weight"])

global_min_weight = -0.0158469807356596
global_max_weight = 0.13336576521396637

# 히트맵 생성
plt.figure(figsize=(10, 8))
plt.imshow(weights, cmap="coolwarm", aspect="auto", vmin=global_min_weight, vmax=global_max_weight)  # 공통 스케일 적용
plt.colorbar(label="Weight")  # 컬러바 추가

# 축 레이블 설정
plt.xticks(ticks=np.arange(len(set_global_ids)), labels=set_global_ids, rotation=90)
plt.yticks(ticks=np.arange(len(valid_gene_indices)), labels=valid_gene_indices)

# 제목 및 축 레이블 추가
plt.title(f"{cell_line}-{drug}-{pathway} Weight Heatmap")
plt.xlabel("Drug Substructures")
plt.ylabel("Gene Indices")

# 히트맵 표시
plt.tight_layout()
plt.show()
```
